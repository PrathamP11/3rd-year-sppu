```python
#finding outlier
from scipy import stats
dfn = pd.read_csv("Desktop/dsbda code/demoa2.csv")
```

```python
dfn
```

```python
z = np.abs(stats.zscore(dfn['math score']))
```

```python
print(z)
```

```python
threshold = 0.18
```

```python
sample_outlier = np.where (z<threshold)
sample_outlier
```

```python
#removing outlier
new_df=dfn
for i in sample_outlier:
    new_df.drop(i,inplace=True)
new_df
```

```python
#data normalization
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()

x=new_df.iloc[:,:4]

x_scaled = min_max_scaler.fit_transform(x)

df_normalized = pd.DataFrame(x_scaled)

df_normalized
```
